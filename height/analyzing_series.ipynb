```python
import pandas as pd
import zipfile
import json
import numpy as np
from scipy.signal import filtfilt
```

```python
# Replace 'your_zipfile.zip' with the path to your ZIP file
# and 'your_json_file.json' with the JSON file's name inside the ZIP
zip_file_path = 'squats_jumps/20221025T093638_U001_SQUAT_JUMP.zip'
json_file_name = '20221025T093638_U001_SQUAT_JUMP.json'

# Open the ZIP file
with zipfile.ZipFile(zip_file_path, 'r') as z:
    # Read the JSON file content into memory
    with z.open(json_file_name) as f:
        json_data = json.loads(f.read())
```

```python
json_data.keys()
```

```python
df = pd.DataFrame(json_data['data'])
df.head()
```

```python
window_size = 5

# Create a moving average (MA) filter
b = np.ones(window_size) / window_size
# Apply the filter in forward and reverse directions
df['linAccZ_Smooth'] = filtfilt(b, 1, df['linAccZ'])
df
```

```python
import plotly.express as px

fig = px.line(df[100:1100], y='linAccZ_Smooth', title='Linear Acceleration Z over Time')
fig.show()
```
